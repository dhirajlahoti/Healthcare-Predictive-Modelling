```python
import pandas as pd

#converting the dataset from .xpt to .csv format
xpt_data = pd.read_sas("LLCP2022.XPT") #enter the path where the LLCP2022.XPT file is stored 

# Save as CSV
xpt_data.to_csv('LLCP2022.csv', index=False) #enter the path where you wish to save the csv version
xpt_data.shape
```

```python
# select required columns
xpt_data_selected = xpt_data[['DIABETE4', 'SEXVAR', '_AGEG5YR', '_BMI5', 'COVIDPOS', 'SMOKE100',
                              '_TOTINDA', 'SDHFOOD1', 'ALCDAY4', 'AVEDRNK3','MEDCOST1', 'GENHLTH','INCOME3', 'CVDSTRK3', 'DIFFWALK','_MICHD']]
#Drop null Values
xpt_data_selected = xpt_data_selected.dropna()
xpt_data_selected.shape
```

```python
#calculating the value of number of drinks consumed in a month
def new_column_calculate(row):
    if row['ALCDAY4'] > 200:
        return row['ALCDAY4'] - 200
    else:
        return (row['ALCDAY4'] - 100)*4

xpt_data_selected['Alcohol_consumed'] = xpt_data_selected.apply(new_column_calculate, axis=1)
del xpt_data_selected['ALCDAY4']
```

```python
# Change 2 to 0 because it is No
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)

xpt_data_selected['DIFFWALK'] = xpt_data_selected['DIFFWALK'].replace({2:0})
xpt_data_selected = xpt_data_selected[xpt_data_selected.DIFFWALK != 7]
xpt_data_selected = xpt_data_selected[xpt_data_selected.DIFFWALK != 9]
```

```python
# 0 is for no diabetes or only during pregnancy, 1 is for pre-diabetes or borderline diabetes, 2 is for yes diabetes
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)
xpt_data_selected['DIABETE4'] = xpt_data_selected['DIABETE4'].replace({2:0, 3:0, 1:2, 4:1})
xpt_data_selected = xpt_data_selected[xpt_data_selected.DIABETE4 != 7]
xpt_data_selected = xpt_data_selected[xpt_data_selected.DIABETE4 != 9]
```

```python
#BMI (the values are BMI * 100. So for example a BMI of 3245 actually translates to 32.45)
xpt_data_selected['_BMI5'] = xpt_data_selected['_BMI5'].div(100)
```

```python
# Change 2 to 0 because it is No
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)
xpt_data_selected['SMOKE100'] = xpt_data_selected['SMOKE100'].replace({2:0})
xpt_data_selected = xpt_data_selected[xpt_data_selected.SMOKE100 != 7]
xpt_data_selected = xpt_data_selected[xpt_data_selected.SMOKE100 != 9]
```

```python
# Change 2 to 0 because it is negative
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)

xpt_data_selected['COVIDPOS'] = xpt_data_selected['COVIDPOS'].replace({2:0})
xpt_data_selected = xpt_data_selected[xpt_data_selected.COVIDPOS != 7]
xpt_data_selected = xpt_data_selected[xpt_data_selected.COVIDPOS != 9]
```

```python
# Change 2 to 0 because it means No
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)

xpt_data_selected['SDHFOOD1'] = xpt_data_selected['SDHFOOD1'].replace({2:0})
xpt_data_selected = xpt_data_selected[xpt_data_selected.SDHFOOD1 != 7]
xpt_data_selected = xpt_data_selected[xpt_data_selected.SDHFOOD1 != 9]
```

```python
# Change 2 to 0 because it is No
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)
xpt_data_selected['CVDSTRK3'] = xpt_data_selected['CVDSTRK3'].replace({2:0})
xpt_data_selected = xpt_data_selected[xpt_data_selected.CVDSTRK3 != 7]
xpt_data_selected = xpt_data_selected[xpt_data_selected.CVDSTRK3 != 9]

xpt_data_selected['_MICHD'] = xpt_data_selected['_MICHD'].replace({2: 0})
```

```python
# Change 2 to 0 because it is No
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)

xpt_data_selected['_TOTINDA'] = xpt_data_selected['_TOTINDA'].replace({2:0})
xpt_data_selected = xpt_data_selected[xpt_data_selected._TOTINDA != 9]
```

```python
# Change 2 to 0 because it is No
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)
xpt_data_selected['MEDCOST1'] = xpt_data_selected['MEDCOST1'].replace({2:0})
xpt_data_selected = xpt_data_selected[xpt_data_selected.MEDCOST1 != 7]
xpt_data_selected = xpt_data_selected[xpt_data_selected.MEDCOST1 != 9]
```

```python
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)
xpt_data_selected = xpt_data_selected[xpt_data_selected.GENHLTH != 7]
xpt_data_selected = xpt_data_selected[xpt_data_selected.GENHLTH != 9]
```

```python
# Remove all values that are 7 (dont knows)
# Remove all values that are 9 (refused to answer)
xpt_data_selected = xpt_data_selected[xpt_data_selected.INCOME3 != 77]
xpt_data_selected = xpt_data_selected[xpt_data_selected.INCOME3 != 99]
# this step removed a lot of values
```

```python
xpt_data_selected['SEXVAR'] = xpt_data_selected['SEXVAR'].replace({2:0})

# Value category 14 refers to the people who refused to answer
xpt_data_selected = xpt_data_selected[xpt_data_selected._AGEG5YR != 14]
```

```python
# Converting the categorical variables from numeric values to string

xpt_data_selected = xpt_data_selected.astype(str)
result = xpt_data_selected.dtypes

xpt_data_selected['_BMI5'] = xpt_data_selected['_BMI5'].astype(float)

xpt_data_selected['AVEDRNK3'] = xpt_data_selected['AVEDRNK3'].astype(float)
xpt_data_selected['Alcohol_consumed'] = xpt_data_selected['Alcohol_consumed'].astype(float)
print(xpt_data_selected.dtypes)
xpt_data_selected.describe(include='object')
```

```python
# Renaming features to make them more readable and performing some basic descriptive analytics
xpt_data_selected = xpt_data_selected.rename(columns={'DIABETE4':'Diabetes', 
                                         'SEXVAR':'Sex',  
                                         '_AGEG5YR':'Age',
                                         '_BMI5':'BMI', 
                                         'SMOKE100':'Smoker', 
                                         'CVDSTRK3':'Stroke', '_MICHD':'HeartDiseaseorAttack', 
                                         '_TOTINDA':'Physical_activity', 
                                         'SDHFOOD1':'Food_shortage', 'INCOME3':"Income_category", 
                                         'DIFFWALK':'Walking_difficulty', 'GENHLTH': 'General_health',
                                         'AVEDRNK3': 'Average_drink' 
                                         })
xpt_data_selected.shape
df = pd.DataFrame(xpt_data_selected)
summary_BMI = df['BMI'].describe()
print(summary_BMI)

diabetes_count = df['Diabetes'].value_counts()
diabetes_percentage = (diabetes_count/len(df))*100
summary_diabetes = pd.DataFrame({'Diabetes_Count': diabetes_count, 'Diabetes_percentage': diabetes_percentage})
print(summary_diabetes)

smoker_count = df['Smoker'].value_counts()
smoker_percentage = (smoker_count/len(df))*100
summary_smoker = pd.DataFrame({'Smoker_Count': smoker_count, 'Smoker_percentage': smoker_percentage})

print(summary_smoker)

heart_disease_count = df['HeartDiseaseorAttack'].value_counts()
heart_disease_count_percentage = (heart_disease_count/len(df))*100
summary_heart_disease = pd.DataFrame({'heart_disease_count': heart_disease_count, 'heart_disease_percentage': heart_disease_count_percentage})

print(summary_heart_disease)
xpt_data_selected.to_csv('filtered_data.csv')
```
